
<p><h3><b><font color = 'DarkBlue'> Load the MNIST dataset </font></b></h3></p>

```python
from keras.datasets import mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape
```

```python
train_labels.shape
```


<p><h3><b><font color = 'DarkBlue'> Preprocess the train and test images </font></b></h3></p>

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28)) 
test_images = test_images.astype('float32') / 255
```


<p><h3><b><font color = 'DarkBlue'> Creating train and test labels for identifying digit 2.  </font></b></h3></p>

The label is set to 1 for digit 2 otherwise set to 0. 

```python
train_labels[:20]
```

```python
import numpy as np
```

```python
train_labels_new = np.where(train_labels==2,1,0)
```

```python
print(train_labels[:20])
print(train_labels_new[:20])
```

```python
test_labels[:20]
```

```python
test_labels_new = np.where(test_labels==2,1,0)
```

```python
print(test_labels[:20])
print(test_labels_new[:20])
```


<p><h3><b><font color = 'DarkBlue'> Converting the train and test labels to categorical  </font></b></h3></p>

This is required to train the model_A which is created and trained for identifying digits from 0 to 9. 

```python
from keras.utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels[:10]
```


<p><h3><b><font color = 'DarkBlue'> Building a model to train MNIST data (0-9) </font></b></h3></p>

```python
from keras import models
from keras import layers
```

```python
model_A = keras.models.Sequential([
 keras.layers.Dense(units=500, input_shape=(28 * 28,), activation='relu'),
 keras.layers.Dense(units=300, activation='relu'),
 keras.layers.Dense(units=300, activation='relu'),
 keras.layers.Dense(units=200, activation='relu'),
 keras.layers.Dense(units=100, activation='relu'),
 keras.layers.Dense(units=10, activation='softmax'),
 ])
```

```python
model_A.summary()
```

```python
model_A.compile(loss='categorical_crossentropy', optimizer = keras.optimizers.SGD(learning_rate=0.01), metrics=['accuracy'])
```

```python
model_A.fit(train_images, train_labels,epochs=10,batch_size=512,validation_data=(test_images,test_labels))
```


<p><h3><b><font color = 'DarkBlue'> Save model_A for reuse </font></b></h3></p>

```python
import os
os.getcwd()
model_A.save("trained_model_A.h5")
```


<p><h3><b><font color = 'DarkBlue'> Load model_A and create a clone for reuse </font></b></h3></p>

```python
model_pretrained = keras.models.load_model("trained_model_A.h5")
```

```python
model_A_clone = keras.models.clone_model(model_pretrained)
model_A_clone.set_weights(model_pretrained.get_weights())
```


<p><h3><b><font color = 'DarkBlue'> Build a new model using all the layers of model_A with a new output layer for classifying digit 2 </font></b></h3></p>

```python
model_new = keras.models.Sequential(model_A_clone.layers[:-1])
model_new.add(keras.layers.Dense(1, activation="sigmoid"))
```

```python
model_new.summary()
```

```python
for layer in model_new.layers[:-1]:         # freeze the reused layers for some epochs giving the new layer some time to learn reasonable weights. 
  layer.trainable = False
```


<p><h3><b><font color = 'DarkBlue'> Compile new model </font></b></h3></p>

we must compile the model after freeze or unfreeze layers.


```python
model_new.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

<p><h3><b><font color = 'DarkBlue'> Fit the new model for classifying digit 2 </font></b></h3></p>

Train model_new for the classifying digit 2 first for few epochs to avoid error. 

```python
history = model_new.fit(train_images, train_labels_new, epochs=5, batch_size=512, validation_data=(test_images,test_labels_new))
```

```python
for layer in model_new.layers[:-1]:      # unfreeze the reused layers  
  layer.trainable = True
```

```python
model_new.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
model_new.fit(train_images, train_labels_new, epochs=15, batch_size=512, validation_data=(test_images,test_labels_new))   # fit the new model on the data
```

```python
model_new.evaluate(test_images, test_labels_new)      # Evaluate the model on test dataset 
```

```python
y_pred = model_new.predict_classes(test_images)   # predict_classes() outputs the class with highest estimated probability
y_pred=y_pred.flatten()                           # Convert it to 1D array
y_pred                                            # new predicted labels for image 2 as 1 otherwise 0
```
